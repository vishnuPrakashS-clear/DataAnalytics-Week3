```python
import pandas as pd
import numpy as np

path = "/content/Data-cleaning-for-beginners-using-pandas.csv"
df = pd.read_csv(path)
df.style.hide_index()
```

# copy of the dataset


```python
df_copy = df
df_copy.style.hide_index()
```

#Missing Values


```python
df_copy.isnull().sum()
```

Data Types


```python
df_copy.dtypes
```

Outliers


```python
df_outlier = df_copy[["Age", "Rating"]]
df_outlier.describe()
```

Salary Formatting


```python
df_copy['Salary'] = df_copy['Salary'].str.replace("$", " ")
df_copy['Salary'] = df_copy['Salary'].str.replace("k","000")
df_copy['Salary'] = df_copy.Salary
print(df_copy)
```

Location Standardisation


```python
df_copy['Location'] = df_copy['Location'].replace({ 'New York,Ny': 'New York', 'India In': 'India', 'India,In': 'India','Australia Aus': 'Australia'})
df_copy['Location']
```

Established Column


```python
df_copy['Established'] = df_copy['Established'].replace(-1, np.nan)
df_copy
```

```python
df_copy['Established'] = df_copy['Established'].replace(np.nan, df_copy['Established'].mode().iloc[0])
df_copy
```

```python
df_copy['Established'] = df_copy['Established'].astype(int)
df_copy
```

Easy Apply


```python
df_copy.rename(columns={"Easy Apply": "Apply"}, inplace=True)
df_copy['Apply'] = df_copy['Apply'].replace("-1","FALSE")
df
```

Rating Range


```python
df_copy['Rating'].fillna(df_copy['Rating'].mean(), inplace = True)
df_copy['Rating'] = df_copy.Rating.round(decimals = 1)
df
```

```python
df_copy['Rating'] = df_copy['Rating'].abs()
df
```

```python
avg_age = df['Age'].mean()

df['Age']= df['Age'].fillna(avg_age)
df['Age']= df['Age'].round(decimals = 1)
df
```
